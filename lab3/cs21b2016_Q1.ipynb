```python
import numpy as np
import pandas as pd 
import scipy as stats
```

### loading dataset

```python
dataset = pd.read_csv('iris.csv')  

X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

display(X)
display(y)
```

## Choose one random sample from each class (Setosa, Versicolor, and Virginica) which will act as the test data.

```python

# Separate data into different classes
setosa_indices = np.where(y == 0)[0]
versicolor_indices = np.where(y == 1)[0]
virginica_indices = np.where(y == 2)[0]


# Choose one random sample from each class for X_test
setosa_test_index = np.random.choice(setosa_indices)
versicolor_test_index = np.random.choice(versicolor_indices)
virginica_test_index = np.random.choice(virginica_indices)

# Combine the chosen indices for X_test
X_test_indices = [setosa_test_index, versicolor_test_index, virginica_test_index]
print("X_test_indices",X_test_indices)

# Create the X_test and y_test sets
X_test = X[X_test_indices]
y_test = y[X_test_indices]

# Create the X_train and y_train sets using the remaining data
X_train_indices = np.setdiff1d(np.arange(len(X)), X_test_indices)
X_train = X[X_train_indices]
y_train = y[X_train_indices]


print("X_train",X_train)
print("y_train",y_train)

print("X_test",X_test)
print("y_test",y_test)

# Print the shapes of the sets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

## Compute the mean vector and covariance matrix for each class in the X_train

```python

# Calculate the Mahalanobis distance for each sample in test
mean_vector_setosa = np.mean(X_train[y_train == 0], axis=0)
print("mean_vector of setosa:\n",mean_vector_setosa)
cov_matrix_setosa = np.cov(X_train[y_train == 0], rowvar=False)
print("covariance_mat of setosa:\n",cov_matrix_setosa)
inv_cov_matrix_setosa = np.linalg.inv(cov_matrix_setosa)
print("inv_cov of setosa:\n",inv_cov_matrix_setosa)

print("\n\n\n")

mean_vector_versicolor= np.mean(X_train[y_train == 1], axis=0)
print("mean_vector of versicolor:\n",mean_vector_versicolor)
cov_matrix_versicolor = np.cov(X_train[y_train == 1], rowvar=False)
print("covariance_mat of versicolor:\n",cov_matrix_versicolor)
inv_cov_matrix_versicolor = np.linalg.inv(cov_matrix_versicolor)
print("inv_cov of versicolor:\n",inv_cov_matrix_versicolor)

print("\n\n\n")

mean_vector_virginica = np.mean(X_train[y_train == 2], axis=0)
print("mean_vector of virginica:\n",mean_vector_virginica)
cov_matrix_virginica = np.cov(X_train[y_train == 2], rowvar=False)
print("covariance_mat of virginica:\n",cov_matrix_virginica)
inv_cov_matrix_virginica = np.linalg.inv(cov_matrix_virginica)
print("inv_cov of virginica:\n",inv_cov_matrix_virginica)

```

## calculation of  mahalanobis distance of 'x_test' samples to corresponding distance to each class

```python
matrix = []
mahalanobis_setosa = []
mahalanobis_versicolor = []
mahalanobis_virginica = []

for row in X_test:
    diff = row - mean_vector_setosa
    mahalanobis_dist = np.sqrt(np.dot(np.dot(diff, inv_cov_matrix_setosa),diff.T))
    mahalanobis_setosa.append(mahalanobis_dist)

matrix.append(mahalanobis_setosa)

for row in X_test:
    diff = row - mean_vector_versicolor
    mahalanobis_dist = np.sqrt(np.dot(np.dot(diff, inv_cov_matrix_versicolor),diff.T))
    mahalanobis_versicolor.append(mahalanobis_dist)

matrix.append(mahalanobis_versicolor)

for row in X_test:
    diff = row - mean_vector_virginica
    mahalanobis_dist = np.sqrt(np.dot(np.dot(diff, inv_cov_matrix_virginica),diff.T))
    mahalanobis_virginica.append(mahalanobis_dist)

matrix.append(mahalanobis_virginica)


# Define row names
row_names = ['setosa_mahalanobis', 'versicolor_mahalanobis', 'virginica_mahalanobis']

# Define column names
column_names = ['test_sample_1', 'test_sample_2', 'test_sample_3']

# Create a DataFrame from the matrix
df = pd.DataFrame(matrix, columns=column_names, index=row_names)

# Print the DataFrame(df)
display(df)
```

## Compare the Mahalanobis distances for the three samples and classify each sample to the class with the smallest Mahalanobis distance.


```python


# Initialize a list to store the indices of minimum values in each column
y_pred = []
maha = []

# Get the number of columns in the matrix
num_columns = len(matrix[0])

for col in range(num_columns):
    min_value = matrix[0][col]
    min_index = 0

    # Loop through each row in the column
    for row in range(1, len(matrix)):
        if matrix[row][col] < min_value:
            min_value = matrix[row][col]
            min_index = row
    
    y_pred.append(min_index)
    maha.append(min_value)

# Print the indices of minimum values in each column
print("Indices of minimum values in each column (y_pred) :", y_pred)
```

## Print the original class and the predicted class for each sample, along with their Mahalanobis distances.


```python
mat =[]
# because y_test will be like [0 1 2] we want to convert like [0,1,2]
new_ytest = [y_test[0],y_test[1],y_test[2]]

mat.append(new_ytest)
mat.append(y_pred)
mat.append(maha)

# Define column names
column_names = ['test_sample_1', 'test_sample_2', 'test_sample_3']

# Define column names
row_names = ['y_test', 'y_pred', 'mahalanobis_distance']

# Create a DataFrame from the matrix
df1 = pd.DataFrame(mat, columns=column_names,index = row_names)

# Print the DataFrame(df)
display(df1)
```

## accuracy of your classifier on the testing set and report the results.

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
print("percent accuracy score : ",accuracy_score(y_test, y_pred)*100)
```

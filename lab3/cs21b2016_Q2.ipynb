## convert table to dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# list of name, degree, score
hr_study =[2,3,4,5,6,7]
hr_tv = [4,3,2,1,0,0]
hr_activity = [2,4,6,8,10,12]
hr_music =[2,3,4,1,5,0]
water_consume = [5,6,5,6,4,5] 
test_Score = [65,70,75,80,85,90]
# dictionary of lists
dict = {'hr_study': hr_study, 'hr_tv': hr_tv, 'hr_activity': hr_activity,'hr_music': hr_music,'water_consume':water_consume,'test_Score': test_Score}
     
dataset = pd.DataFrame(dict)
 
display(dataset)
```

## Calculate correlation coefficients

```python
# Calculate means
mean_test_score = sum(test_Score) / len(test_Score)


correlations = {}
for col in dataset.columns:
    if col != 'test_Score':
        mean_col = sum(dataset[col]) / len(dataset[col])
        numerator = sum((dataset[col] - mean_col) * (dataset['test_Score'] - mean_test_score))
        denominator = (sum((dataset[col] - mean_col) ** 2) * sum((dataset['test_Score'] - mean_test_score) ** 2)) ** 0.5
        correlation = numerator / denominator
        correlations[col] = correlation

print(correlations,"\n\n")
# Interpret correlations
for col, correlation in correlations.items():
    if correlation > 0.5:
        print(f"{col} has a positive correlation with test_Score: {correlation:.2f}")
    elif correlation < -0.5:
        print(f"{col} has a negative correlation with test_Score: {correlation:.2f}")
    else:
        print(f"{col} has no significant correlation with test_Score: {correlation:.2f}")
```

```python
# Scatter plot for Hours of Study vs Test Score
plt.figure(figsize=(6, 4))
plt.scatter(dataset['hr_study'], dataset['test_Score'])
plt.title("Hours of Study vs Test Score")
plt.xlabel("Hours of Study")
plt.ylabel("Test Score")
plt.grid(True)
plt.show()
```

Opportunity Cost: Every hour spent watching TV could potentially be an hour that could have been spent studying or engaging in other productive activities related to academics. This concept of opportunity cost highlights that time is a limited resource, and how it's allocated can significantly impact outcomes.

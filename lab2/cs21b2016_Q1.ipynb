```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### importing dataset


```python
dataset = pd.read_csv("iris.csv")
X = dataset.iloc[:,1:3].values
y = dataset.iloc[:,-1].values

print(X)
print(y)
```

## splitting the dataset into training and test set


```python
X_train = dataset.iloc[3:108,1:3].values
y_train = dataset.iloc[3:108,-1].values

X_test = dataset.iloc[0:3,1:3].values
y_test = dataset.iloc[0:3,-1].values

print(X_train)
print(y_train)

print(X_test)
print(y_test)
```

## encoding the dependent variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)

print(y_train)
print(y_test)
```

## plotting training samples

```python
plt.scatter(X_train[y_train == 0, 0], X_train[y_train == 0, 1], s = 50, c = 'red', label = 'Iris-setosa')
plt.scatter(X_train[y_train == 1, 0], X_train[y_train == 1, 1], s = 50, c = 'blue', label = 'Iris-versicolor')
plt.scatter(X_train[y_train == 2, 0], X_train[y_train == 2, 1], s = 50, c = 'green', label = 'Iris-virginica')
plt.title('plotting training samples')
plt.xlabel('sepel-length')
plt.ylabel('sepal-width')
plt.legend()
plt.show()
```

## calculating  k-samples distances and finding majority

```python

import math
y_pred = []

def classifyAPoint(points,p,k=3):

	distance=[]
	for group in points:
		for feature in points[group]:

			#calculate the euclidean distance of p from training points
			euclidean_distance = math.sqrt((feature[0]-p[0])**2 +(feature[1]-p[1])**2)

			# Add a tuple of form (distance,group) in the distance list
			distance.append((euclidean_distance,group))

	distance = sorted(distance)[:k]
	

	freq1 = 0 #frequency of group 0
	freq2 = 0 #frequency of group 1
	freq3 = 0 #frequency of group 2

	for d in distance:
		if d[1] == 0:
			freq1 += 1
		elif d[1] == 1:
			freq2 += 1
		elif d[1] == 2:
			freq2 += 1

	if freq1>=freq2>=freq3 or freq1>=freq3>=freq2:
		return 0
	elif freq2>=freq1>=freq3 or freq2>=freq3>=freq1:
		return 1
	else:
		return 2


# driver function
def main():


	points = {0:X_train[y_train == 0],
				1:X_train[y_train == 1],
				2:X_train[y_train == 2]}

	# testing point p(x,y)
	# Number of neighbours
	k = 3

	for i in range(3): 
		p = X_test[i]
		print(p)
		print("The value classified to above point is: {}\n".\
		format(classifyAPoint(points,p,k)))

		y_pred.append(classifyAPoint(points,p,k))
	print("y-predicted matrix\n",y_pred)




if __name__ == '__main__':
	main()
	
	
# This code is contributed by Atul Kumar (www.fb.com/atul.kr.007)
```

## accuracy of your classifier on the testing set and report the results.


```python
print("y-predicted matrix\n",y_pred)
print("y-test matrix\n",y_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
print("percent accuracy score : ",accuracy_score(y_test, y_pred)*100)
```

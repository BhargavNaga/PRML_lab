```python
def append_record(df2):
    print(df2)
    nme = input("enter name")
    ac_no = int(input("enter account no"))
    ac_type = input("enter account type")
    aadhar_no = int(input("enter aadhar no"))
    balance = int(input("enter balance"))
# dictionary of lists
    dict = {'name': [nme], 'Account_no': [ac_no], 'Account_type': [ac_type],'Aadhar_no': [aadhar_no],'Balance':[balance] }

    new_row = pd.DataFrame(dict)

    df2 = pd.concat([new_row,df],ignore_index = True)
    display(df2)
    df2.to_csv("SBIAccountHolder.csv",index=False)


def delete_record(df2):
  acc_num=int(input("Enter Account Number to delete record : "))
  #df2=df2[df2["Account Number"]!=acc_num]
  df2.drop(df2.loc[df2["Account_no"]==acc_num].index,inplace=True)
  display(df2)
  df2.to_csv("SBIAccountHolder.csv",index=False)

def credit(df2):
  acc_num=int(input("Enter Account Number to credit amount : "))
  amt =int(input("Enter amount to be credited : "))
  df2.loc[df2.loc[df2["Account_no"]==acc_num].index,"Balance"] += amt
  display(df2)
  df2.to_csv("SBIAccountHolder.csv",index=False)

def debit(df2):
   acc_num = int(input("Enter Account Number to debit amount : "))
   amt = int(input("Enter amount to be debited : "))
   pos= df2.loc[df2["Account_no"]==acc_num]
   if((pos["Balance"]>=amt).bool()) :
        #df2.loc[df2.loc[df2["Account Number"]==acc_num].index,"Balance"] -= amt
        #df2[,4]=pos["Balance"]
        pos["Balance"]-=amt
        df2.loc[df2["Account_no"]==acc_num]=pos
   else :
        print("Insufficent Balance")
   print(df2)
   df2.to_csv("SBIAccountHolder.csv",index=False)

def print_details(df2):
   acc_num=int(input("Enter Account Number to print details : "))
   pos=df2.loc[df2["Account_no"]==acc_num]
   print(pos)

if __name__ =="__main__":

    print("1.Append Record (row) of account holder ")
    print("2.Delete Record, given the account number")
    print("3.Credit : Ask the user to enter the amount to be credited, and then add it to the balance.")
    print("4.Debit : Ask the user to enter the amount to be debited, say x, and then subtract it from the balance. In case of SB account type, debit operation should report an error if (balance-x) &lt; 0")
    print("5.Print Account details given the account no")
    print("enter other than 1-5  character to exit")




    while(1):
        choice = int(input("enter choice"))

        if(choice == 1):
            df2 = pd.read_csv("SBIAccountHolder.csv") 
            append_record(df2)
        elif(choice == 2):
            df2 = pd.read_csv("SBIAccountHolder.csv") 
            delete_record(df2)
        elif(choice == 3):
            df2 = pd.read_csv("SBIAccountHolder.csv") 
            credit(df2)
        elif(choice == 4):
            df2 = pd.read_csv("SBIAccountHolder.csv") 
            debit(df2)
        elif(choice == 5):
            df2 = pd.read_csv("SBIAccountHolder.csv") 
            print_details(df2)
        else:
            print("---exit--- ")
            break




   




```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## converting images from BGR TO RGB format

```python
#cs21b2016_Q2
query_img = cv2.imread("queryimage.png")
ref_img1 = cv2.imread("refimage1.png")
ref_img2 = cv2.imread("refimage2.png")
img = cv2.cvtColor(query_img,cv2.COLOR_BGR2RGB)
ref1 = cv2.cvtColor(ref_img1,cv2.COLOR_BGR2RGB)
ref2 = cv2.cvtColor(ref_img2,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
plt.imshow(ref1)
plt.show()
plt.imshow(ref2)
plt.show()
```

## Calculate histograms for each image

```python
# Calculate histograms for each image
hist_reference1 = cv2.calcHist([ref1], [0], None, [256], [0, 256])
hist_reference2 = cv2.calcHist([ref2], [0], None, [256], [0, 256])
hist_query = cv2.calcHist([img], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.title("Reference1 Histogram")
plt.plot(hist_reference1)
plt.xlim([0, 256])

plt.subplot(2, 3, 2)
plt.title("Reference2 Histogram")
plt.plot(hist_reference2)
plt.xlim([0, 256])

plt.subplot(2, 3, 3)
plt.title("Query Histogram")
plt.plot(hist_query)
plt.xlim([0, 256])
```

## Normalize histograms for each image

```python
# Normalize histograms
hist_reference1 /= hist_reference1.sum()
hist_reference2 /= hist_reference2.sum()
hist_query /= hist_query.sum()

# Plot histograms
plt.figure(figsize=(12, 6))

# Plot normalized histograms
plt.subplot(2, 3, 4)
plt.title("Reference1 Normalized Histogram")
plt.plot(hist_reference1)
plt.xlim([0, 256])

plt.subplot(2, 3, 5)
plt.title("Reference2 Normalized Histogram")
plt.plot(hist_reference2)
plt.xlim([0, 256])

plt.subplot(2, 3, 6)
plt.title("Query Normalized Histogram")
plt.plot(hist_query)
plt.xlim([0, 256])

```

## calculate Bhattacharyya Distance

```python

bhattacharyya_distance1 = cv2.compareHist(hist_reference1, hist_query, cv2.HISTCMP_BHATTACHARYYA)
bhattacharyya_distance2 = cv2.compareHist(hist_reference2, hist_query, cv2.HISTCMP_BHATTACHARYYA)

print(f"Bhattacharyya Distance between Reference1 and Query: {bhattacharyya_distance1}")
print(f"Bhattacharyya Distance between Reference2 and Query: {bhattacharyya_distance2}")
```
